# Movie Recommendation system

```python
credits_df=pd.read_csv('tmdb_5000_credits.csv')
```

```python
# Read the credit dataset
credits_df.head()
```

```python
movies_df= pd.read_csv('tmdb_5000_movies.csv')
```

```python
# Reading the movies dataset
movies_df.head()
```

```python
# Find the size (No. of rows and columns) in credit and movies dataset
print("Credits:", credits_df.shape)
print("movies:", movies_df.shape)
```

### Step 1: Merging the two datasets with the common feature of movie id

```python
# Rename the movie_id feature as id in credit_df dataset
credits_rename = credits_df.rename(index=str, columns={'movie_id':'id'})
```

```python
# Merge two dataset with common id
movies_df_merge=movies_df.merge(credits_rename, on='id')
```

```python
movies_df_merge.head()
```

```python
print("movies_df_merge:", movies_df_merge.shape)
```

### Step:2 Remove the unwanted features

```python
movies_df_cleaned=movies_df_merge.drop(columns=['title_x','title_y','status','production_countries','homepage'])
movies_df_cleaned.head()
```

```python
print("movies_df_cleaned:", movies_df_cleaned.shape)
```

### Step 3: Check for NAN values

```python
movies_df_cleaned.info()
```

```python
# There are no features with the NAN values so continue with the existing dataset
```

## Recommendation system based on Rating

### Step 4: Calculating Weighted Average Rating of each movie

### Using formula  W=(R×v+C×m)/(v+m)
    where W- weighted rating
          R- Average rating for the movies from 0 to 10 
          v- Number of votes for the movies
          C- Mean vote across the whole record
          m- Minimum vote required to get listed

```python
#Calculation of each parameter in the given formula
R=movies_df_cleaned['vote_average']
v=movies_df_cleaned['vote_count']
C=movies_df_cleaned['vote_average'].mean()
m=movies_df_cleaned['vote_count'].quantile(0.70)
```

```python
movies_df_cleaned['weighted_average']=(((R*v)+(C*m))/(v+m))
```

```python
movies_df_cleaned.head()
```

### Step 5: Movies are ranked based on weighted average rating

```python
# Based on the weighted average rating, the movies are sorted
movies_df_ranking=movies_df_cleaned.sort_values('weighted_average', ascending =False)
```

```python
movies_df_ranking[['original_title','vote_average','vote_count','weighted_average','popularity']].head(20)
```

### Step 6: Rating based recommended movies are plotted 

```python
plt.figure(figsize=(12,6))
sns.barplot(x=movies_df_ranking['weighted_average'].head(10),y=movies_df_ranking['original_title'].head(10))
plt.title('Best Movies by Average Votes', weight='bold')
plt.xlabel('Weighted Average Score', weight='bold')
plt.ylabel('Movie Title', weight='bold')
```

## Recommendation system based on popularity

###  Step 7: Movies are ranked based on Popularity

```python
movies_df_popularity=movies_df_cleaned.sort_values('popularity', ascending=False)
```

```python
movies_df_popularity[['original_title','weighted_average','popularity']].head(20)
```

### Step 8: Popularity based recommended movies are plotted

```python
plt.figure(figsize=(12,6))
sns.barplot(x=movies_df_popularity['popularity'].head(10),y=movies_df_popularity['original_title'].head(10))
plt.title('Best Movies by Popularity',weight='bold')
plt.xlabel('Popularity', weight='bold')
plt.ylabel('Movie title', weight='bold')
```

## Recommendation system based on both Rating and Popularity

```python
# Equal priority is given to both rating and popularity
```

```python
from sklearn.preprocessing import MinMaxScaler
scalar= MinMaxScaler()
movies_df_scalar=scalar.fit_transform(movies_df_cleaned[['weighted_average','popularity']])
movies_df_normalised=pd.DataFrame(movies_df_scalar, columns=['weighted_average','popularity'])
movies_df_normalised.head()
```

```python
movies_df_cleaned[['normalised_weighted_average','normalised_popularity']]=movies_df_normalised
```

```python
movies_df_cleaned.head()
```

```python
movies_df_cleaned['score']=movies_df_cleaned['normalised_weighted_average']*0.5+movies_df_cleaned['normalised_popularity']*0.5
movies_df_scored=movies_df_cleaned.sort_values(['score'], ascending=False)
movies_df_scored[['original_title','normalised_weighted_average','normalised_popularity','score']].head()
```

```python
plt.figure(figsize=(12,6))
sns.barplot(x=movies_df_cleaned['score'].head(10),y=movies_df_cleaned['original_title'].head(10))
plt.title('Best rated and Best Popular Movies', weight='bold')
plt.xlabel('Score',weight='bold')
plt.ylabel('Movie title', weight='bold')
plt.savefig('Scored movies.png')
```
